```python
# Task 1: Coding Assignment 
# You are given a dataset student_scores.csv containing two columns:
# 1. Hours → Number of study hours per day
# 2. Scores → Marks obtained (out of 100)
# Your task is to perform basic data analysis and visualization.

# Instructions:
# Python Basics
# 1. Load the dataset using pandas.
# 2. Print the first 10 rows.
# 3. Display the number of rows and columns in the dataset.
# 4. Find the maximum, minimum, mean, and median of both Hours and Scores.
# Train-Test Split
# 1. Use train_test_split from sklearn.model_selection to split the dataset into 70% training and 30% testing sets.
# 2. Print the shapes (number of rows) of training and testing sets.
# 3. Ensure reproducibility by setting random_state=42.
# Matplotlib Visualization
# 1. Create a scatter plot of the dataset (Hours vs Scores).
# 2. Highlight training data points in one color and testing data points in another color.
# 3. Add axis labels (Study Hours, Marks), a title, and a legend.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('student_scores.csv')
df.head(10)
```

```python
rows,cols=df.shape
print('No.of rows is',rows)
print('No.of column is',cols)
```

```python
#Maximum and minimum
print("The maximum no. in hours is:",df['Hours'].max())
print("The minimum number in hours is:",df['Hours'].min())
print("The maximum of score is :",df['Scores'].max())
print("The minimum of score is :",df['Scores'].min())

#mean and median
print("The mean of hours",df['Hours'].mean())
print("The mean of scores",df['Scores'].mean())
print("The median of hours",df['Hours'].median())
print("The median of scores",df['Scores'].median())
```

```python
df.isnull().sum()
```

```python
x=df[['Hours']]
y=df[['Scores']]
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
Y_test
```

```python
plt.scatter(df['Hours'],df['Scores'])
```

```python
plt.figure(figsize=(8,5))

plt.scatter(X_train, Y_train, color="green", label="training data")
plt.scatter(X_test, Y_test, color= "blue", label="testing data")

plt.xlabel("study hours")
plt.ylabel("score")
plt.title("study hours vs marks")
plt.legend()
plt.show()
```
